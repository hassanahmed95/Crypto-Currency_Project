## Step 1: Import Libraries and Define Helper Functions

**Purpose:** This section imports all essential Python libraries required for data manipulation, statistical analysis, and visualization throughout the cryptocurrency market analysis.

**What this cell does:**
- Imports `pandas` for data manipulation and DataFrame operations
- Imports `numpy` for numerical computations and mathematical operations
- Imports `matplotlib.pyplot` and `seaborn` for creating professional visualizations
- Configures display settings to suppress warnings and optimize output formatting
- Sets up visualization styles for consistent, professional chart appearance

```python
# Import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Settings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')
sns.set_palette("husl")
%matplotlib inline

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)

print(" All libraries imported successfully!")
```

## Step 2: Load the Data

**Purpose:** Load the cryptocurrency market dataset from CSV file into a pandas DataFrame for analysis.

**What this cell does:**
- Reads the `cryptocurrency.csv` file containing market data for multiple cryptocurrencies
- Displays the total number of rows (data records) loaded
- Shows the column names to understand the data structure
- Previews the first few rows using `head()` to verify successful loading
- Key columns include: timestamp, name, symbol, price_usd, vol_24h, chg_24h, chg_7d, market_cap


```python
# Load cryptocurrency data
df = pd.read_csv('cryptocurrency.csv')

print(f"Dataset loaded: {len(df)} rows")
print(f"Columns: {list(df.columns)}")
print("\nFirst few rows:")
df.head()
```

## Step 3: Clean the Data

**Purpose:** Transform raw data into analysis-ready format by removing formatting characters, converting data types, and handling missing values.

**What this cell does:**
- **Defines `clean_currency()` function:** Removes dollar signs ($) and commas from currency values and converts them to float numbers
- **Defines `clean_percentage()` function:** Removes percentage symbols (%) and plus signs (+) from percentage values and converts to float
- **Applies cleaning functions** to currency columns: price_usd, vol_24h, market_cap
- **Applies cleaning functions** to percentage columns: chg_24h (24-hour change), chg_7d (7-day change)
- **Removes rows with missing critical data** in market_cap, price_usd, or vol_24h columns
- **Removes duplicate entries** by keeping only the latest timestamp for each cryptocurrency symbol
- **Displays data summary** showing data types and cleaned record count
- **Result:** Clean dataset with 239 unique cryptocurrencies ready for analysis


```python
# Function to clean currency values (remove $ and commas)
def clean_currency(value):
    if pd.isna(value):
        return np.nan
    if isinstance(value, str):
        value = value.replace('$', '').replace(',', '')
        try:
            return float(value)
        except:
            return np.nan
    return value

# Function to clean percentage values (remove % and +)
def clean_percentage(value):
    if pd.isna(value):
        return np.nan
    if isinstance(value, str):
        value = value.replace('%', '').replace('+', '')
        try:
            return float(value)
        except:
            return np.nan
    return value

# Clean all columns
print("Cleaning data...")

# Clean currency columns
df['price_usd'] = df['price_usd'].apply(clean_currency)
df['vol_24h'] = df['vol_24h'].apply(clean_currency)
df['market_cap'] = df['market_cap'].apply(clean_currency)

# Clean percentage columns
df['chg_24h'] = df['chg_24h'].apply(clean_percentage)
df['chg_7d'] = df['chg_7d'].apply(clean_percentage)

# Remove rows with missing critical data
df_clean = df.dropna(subset=['market_cap', 'price_usd', 'vol_24h']).copy()

# Remove duplicates (keep the latest entry)
df_clean = df_clean.sort_values('timestamp', ascending=False)
df_clean = df_clean.drop_duplicates(subset=['symbol'], keep='first')

print(f" Data cleaned! {len(df_clean)} unique cryptocurrencies")
print(f"\nData types after cleaning:")
print(df_clean.dtypes)
```

```python
# Quick data summary
print("Data Summary:")
print(df_clean[['price_usd', 'vol_24h', 'market_cap', 'chg_24h', 'chg_7d']].describe())
```

## Step 4: Top 10 Cryptocurrencies by Market Cap

**Purpose:** Identify and rank the largest cryptocurrencies by market capitalization to understand market dominance and established players.

**What this cell does:**
- Uses `nlargest()` to extract the top 10 cryptocurrencies with the highest market capitalization
- Selects relevant columns: name, symbol, price, market cap, 24-hour change, and 7-day change
- Market capitalization = total value of all coins in circulation, indicating the size and stability of each cryptocurrency
- Displays the market leaders that dominate the cryptocurrency ecosystem


```python
# Extract the top 10 cryptocurrencies by market capitalization
top_10_market_cap = df_clean.nlargest(10, 'market_cap')[['name', 'symbol', 'price_usd', 'market_cap', 'chg_24h', 'chg_7d']]

print("Top 10 Cryptocurrencies by Market Cap:")
print("=" * 80)
top_10_market_cap
```

## Step 5: Highest 7-Day Growth Coins

**Purpose:** Identify cryptocurrencies showing the strongest positive momentum over the past week to spot potential growth opportunities.

**What this cell does:**
- Filters dataset to include only coins with market cap > $1 million to exclude extremely small, potentially manipulated coins
- Uses `nlargest()` to find the top 10 cryptocurrencies with highest 7-day percentage growth
- The filter helps avoid "pump and dump" schemes common in very small market cap coins
- Displays coins with the strongest upward price trends, indicating investor interest and momentum
- Useful for identifying coins experiencing significant growth periods


```python
# Identify coins with strongest 7-day growth performance
df_filtered = df_clean[df_clean['market_cap'] > 1_000_000].copy()

# Extract top 10 coins with highest 7-day percentage growth
top_7d_growth = df_filtered.nlargest(10, 'chg_7d')[['name', 'symbol', 'price_usd', 'market_cap', 'chg_7d']]

print("Top 10 Coins by 7-Day Growth (Market Cap > $1M):")
print("=" * 80)
top_7d_growth
```

## Step 6: Highest Volume/Market Cap Ratio

**Purpose:** Identify the most actively traded cryptocurrencies relative to their size, which indicates liquidity and ease of buying/selling.

**What this cell does:**
- Calculates volume-to-market-cap ratio: (24-hour trading volume / market cap) × 100
- A higher ratio means more active trading relative to the coin's size
- High ratios indicate:
  - Strong liquidity (easy to buy/sell without affecting price)
  - Active market participation and investor interest
  - Lower risk of being unable to exit positions
- Identifies top 10 coins with highest trading activity relative to their market capitalization
- Important metric for investors who need to enter/exit positions efficiently


```python
# Calculate volume-to-market-cap ratio as a percentage
df_filtered['vol_to_mcap_ratio'] = (df_filtered['vol_24h'] / df_filtered['market_cap']) * 100

# Identify the top 10 most actively traded coins relative to their size
top_volume_ratio = df_filtered.nlargest(10, 'vol_to_mcap_ratio')[
    ['name', 'symbol', 'vol_24h', 'market_cap', 'vol_to_mcap_ratio']
]

print("Top 10 Coins by Volume/Market Cap Ratio (Most Actively Traded):")
print("=" * 80)
print("Higher ratio = More trading activity relative to size")
print()
print(top_volume_ratio)
```

## Step 7: Most Volatile Coins

**Purpose:** Identify cryptocurrencies with the highest price volatility to understand risk profiles and potential for large price swings.

**What this cell does:**
- Calculates a volatility score using the formula: (|24-hour change| + |7-day change|) / 2
- Takes the absolute value of price changes to measure total volatility regardless of direction
- Higher volatility scores indicate:
  - Greater price instability and risk
  - Potential for larger gains OR losses
  - More suitable for risk-tolerant traders
  - Less suitable for conservative investors
- Identifies top 10 most volatile coins (with market cap > $1M)
- Helps investors match investments to their risk tolerance


```python
# Ensure df_filtered is independent
df_filtered = df_filtered.copy()

# Calculate volatility score (average of absolute changes)
df_filtered['volatility_score'] = (abs(df_filtered['chg_24h']) + abs(df_filtered['chg_7d'])) / 2

# Get top 10 most volatile coins
most_volatile = df_filtered.nlargest(10, 'volatility_score')[
    ['name', 'symbol', 'chg_24h', 'chg_7d', 'volatility_score', 'market_cap']
]

print("Top 10 Most Volatile Coins (Market Cap > $1M):")
print("=" * 80)
print("Higher score = More price volatility")
print()
print(most_volatile)
```

**Purpose:** Discover the relationship between volatility (risk) and growth (reward) to find optimal risk-reward opportunities.

**What this cell does:**
- Creates a sophisticated scatter plot with volatility score on x-axis and 7-day growth on y-axis

## Step 9: Investment Analysis

**Purpose:** Create a comprehensive, multi-factor investment scoring system to rank cryptocurrencies by overall investment attractiveness.

**What this cell does:**
- Filters to coins with market cap > $100M (established cryptocurrencies only)
- **Normalizes multiple metrics to 0-1 scale for fair comparison:**
  - **Growth score (30% weight):** Based on 7-day growth performance
  - **Market cap score (25% weight):** Based on logarithmic market cap (rewards established coins)
  - **Volume score (20% weight):** Based on volume-to-market-cap ratio (liquidity)
  - **Stability score (25% weight):** Inverse of volatility (rewards stable coins)
- **Calculates composite investment score** by combining all four factors with their respective weights
- Uses `nlargest()` to select the top 5 cryptocurrencies with highest investment scores
- **Result:** Data-driven investment recommendations balancing growth, stability, size, and liquidity
- Provides a holistic view beyond just looking at one metric


```python
# Create investment score based on multiple factors
# We want: positive growth, reasonable volatility, good volume, substantial market cap

# Filter coins with market cap > $100M (established coins)
df_invest = df_clean[df_clean['market_cap'] > 100_000_000].copy()

# Normalize metrics (0-1 scale)
df_invest['growth_score'] = (df_invest['chg_7d'] - df_invest['chg_7d'].min()) / (df_invest['chg_7d'].max() - df_invest['chg_7d'].min())
df_invest['mcap_score'] = (np.log10(df_invest['market_cap']) - np.log10(df_invest['market_cap']).min()) / (np.log10(df_invest['market_cap']).max() - np.log10(df_invest['market_cap']).min())
df_invest['volume_score'] = (df_invest['vol_to_mcap_ratio'] - df_invest['vol_to_mcap_ratio'].min()) / (df_invest['vol_to_mcap_ratio'].max() - df_invest['vol_to_mcap_ratio'].min())

# Volatility score (inverse - lower volatility is better)
df_invest['stability_score'] = 1 - ((df_invest['volatility_score'] - df_invest['volatility_score'].min()) / (df_invest['volatility_score'].max() - df_invest['volatility_score'].min()))

# Combined investment score (weighted average)
df_invest['investment_score'] = (
    df_invest['growth_score'] * 0.30 +      # 30% weight on growth
    df_invest['mcap_score'] * 0.25 +        # 25% weight on market cap
    df_invest['volume_score'] * 0.20 +      # 20% weight on volume
    df_invest['stability_score'] * 0.25     # 25% weight on stability
)

# Get top 5 investment recommendations
top_investments = df_invest.nlargest(5, 'investment_score')[['name', 'symbol', 'price_usd', 'market_cap', 
                                                               'chg_7d', 'volatility_score', 'vol_to_mcap_ratio', 
                                                               'investment_score']]

print("Top 5 Investment Recommendations:")
print("=" * 100)
print()
top_investments
```

## Step 10: Detailed Investment Reasoning

**Purpose:** Provide human-readable, comprehensive explanations for each investment recommendation based on their specific characteristics.

**What this cell does:**
- Iterates through the top 5 investment recommendations from Step 9
- For each cryptocurrency, displays:
  - Key metrics: price, market cap, growth, volatility, investment score
  - **Context-aware reasoning based on specific values:**
    - Market cap reasoning (established vs growth potential)
    - Growth reasoning (momentum and trend analysis)
    - Volatility reasoning (risk profile and suitability)
- **Generates dynamic explanations** that adapt to each coin's characteristics
- Provides actionable insights to help investors understand WHY each coin is recommended
- Includes comprehensive disclaimer about risk and the need for independent research
- Makes the quantitative analysis accessible and understandable for all investor types


```python
print("="*100)
print("INVESTMENT RECOMMENDATIONS WITH DETAILED REASONING")
print("="*100)
print()

for idx, row in top_investments.iterrows():
    print(f"🔹 #{top_investments.index.get_loc(idx) + 1}: {row['name']} ({row['symbol'].upper()})")
    print(f"   Price: ${row['price_usd']:,.4f} | Market Cap: ${row['market_cap']:,.0f}")
    print(f"   7-Day Growth: {row['chg_7d']:.2f}% | Volatility: {row['volatility_score']:.2f}")
    print(f"   Investment Score: {row['investment_score']:.3f}")
    print()
    
    # Generate reasoning based on characteristics
    print("   REASONING:")
    
    # Market cap reasoning
    if row['market_cap'] > 1_000_000_000:
        mcap_reason = "This is a well-established cryptocurrency with a market cap over $1B, indicating strong market presence and reduced risk of complete failure."
    elif row['market_cap'] > 100_000_000:
        mcap_reason = "With a market cap over $100M, this coin has established credibility while still offering growth potential."
    else:
        mcap_reason = "This is a smaller market cap coin with higher growth potential, though with increased risk."
    
    # Growth reasoning
    if row['chg_7d'] > 10:
        growth_reason = "The 7-day growth trend is very strong (>10%), showing significant positive momentum in the market."
    elif row['chg_7d'] > 5:
        growth_reason = "Shows solid positive growth over the past week, indicating increasing investor interest and upward price momentum."
    elif row['chg_7d'] > 0:
        growth_reason = "Demonstrates stable positive growth, suggesting steady appreciation and market confidence."
    else:
        growth_reason = "Despite recent negative movement, the overall fundamentals and investment score remain strong."
    
    # Volatility reasoning
    if row['volatility_score'] < 5:
        vol_reason = "Low volatility indicates price stability, making it suitable for risk-averse investors seeking steady returns."
    elif row['volatility_score'] < 10:
        vol_reason = "Moderate volatility provides a balanced risk-reward profile, with reasonable price stability while maintaining growth potential."
    else:
        vol_reason = "Higher volatility suggests more price action, which could present both opportunities and risks for active traders."
    
    print(f"   • {mcap_reason}")
    print(f"   • {growth_reason}")
    print(f"   • {vol_reason}")
    print()
    print("-" * 100)
    print()

print("\n  DISCLAIMER: This analysis is based on historical data and should not be considered financial advice.")
print("    Always do your own research and consult with financial advisors before investing.")
```
